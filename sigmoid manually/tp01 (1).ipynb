```python
import numpy as np

def sigmoid(z):
    return 1 if 1 / (1 + np.exp(-z)) >= 0.5 else 0

def Perceptron(w, b, learning_rate, epochs, X, y):
    for epoch in range(epochs):
        error_moy = 0
        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            y_pred = sigmoid(z)
            error = y[i] - y_pred
            w = w + learning_rate * error * X[i]
            b = b + learning_rate * error
            error_moy += abs(error)

        error_moy /= len(X)
        if error_moy == 0:
            print("Training stopped at epoch:", epoch)
            break

    return w, b

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

w = np.array([1.0, 1.0])
b = 1.0
learning_rate = 0.5
epochs = 100

w, b = Perceptron(w, b, learning_rate, epochs, X, y)

print("final weights:", w)
print("final bias:", b)

correct = sum(sigmoid(np.dot(X[i], w) + b) == y[i]
    for i in range(len(X)))
print(f"Accuracy: {correct}/{len(X)}")

for i in range(len(X)):
    print("prediction for [ X1 =", X[i][0], ", X2 =", X[i][1], "]:", sigmoid(np.dot(X[i], w) + b))
```

```python

def sigmoidRaw(z):
    return 1 / (1 + np.exp(-z))

def sigmoidRslt(z):
    return 1 if sigmoidRaw(z) >= 0.5 else 0

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

np.random.seed(42)
W1 = np.random.randn(2, 2)
b1 = np.random.randn(2)
W2 = np.random.randn(2)
b2 = np.random.randn()

learning_rate = 0.1
epochs = 10000

for _ in range(epochs):
    for i in range(len(X)):
        h = sigmoidRaw(X[i] @ W1 + b1)
        y_pred = sigmoidRaw(h @ W2 + b2)

        error_output = y[i] - y_pred
        delta_output = error_output * y_pred * (1 - y_pred)

        error_hidden = delta_output * W2
        delta_hidden = error_hidden * h * (1 - h)

        W2 += learning_rate * delta_output * h
        b2 += learning_rate * delta_output
        W1 += learning_rate * np.outer(X[i], delta_hidden)
        b1 += learning_rate * delta_hidden

print("Predictions:")
for i in range(len(X)):
    h = sigmoidRaw(X[i] @ W1 + b1)
    pred = sigmoidRslt(h @ W2 + b2)
    print("prediction for [ X1 =", X[i][0], ", X2 =", X[i][1], "]:", pred)

correct = sum(
    sigmoidRslt(sigmoidRaw(X[i] @ W1 + b1) @ W2 + b2) == y[i]
    for i in range(len(X))
)
print(f"Accuracy: {correct}/{len(X)}")
print("Training stopped at epoch:", epochs)
```
